```python
import pandas as pd
import seaborn as sns
import sklearn
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
df = pd.read_csv(r'csv_combined.csv')
```

```python
df.columns
```

```python
df.corr()[['o.c']].sort_values(by='o.c', ascending=False)
plt.figure(figsize=(15, 50))
heatmap = sns.heatmap(abs(df.corr())[['o.c']].sort_values(by='o.c', ascending=False), vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Features Correlating with price close:', fontdict={'fontsize':18}, pad=16);
```
