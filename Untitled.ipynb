```python
df = pd.read_csv(r'csv_combined.csv')
```

```python
df['Time'] = pd.to_datetime(df.Time)
df.set_index('Time')
df.set_index('Time').asfreq('D')
```

```python
train = train.day_close
test = test.day_close
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
train.array.reshape(1, -1)
test.array.reshape(1, -1)
```

```python
scaler.fit(train.array.reshape(-1, 1))
scaled_train = scaler.transform(train.array.reshape(-1, 1))
scaled_test = scaler.transform(test.array.reshape(-1, 1))
```

```python
from keras.preprocessing.sequence import TimeseriesGenerator
```

```python
# define generator
n_input = 7
n_features = 1
generator = TimeseriesGenerator(scaled_train, scaled_train, length=n_input, batch_size=1)
```

```python
X,y = generator[0]
```

```python
X.shape
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
```

```python
# define model
model = Sequential()
model.add(LSTM(100, activation='relu', input_shape=(n_input, n_features)))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.fit(generator,epochs=50)
```

```python
loss_per_epoch = model.history.history['loss']
plt.plot(range(len(loss_per_epoch)),loss_per_epoch)
```

```python
last_train_batch = scaled_train[-7:]
last_train_batch = last_train_batch.reshape((1, n_input, n_features))
model.predict(last_train_batch)
```

```python
scaled_test[0]
```

```python
test_predictions = []

first_eval_batch = scaled_train[-n_input:]
current_batch = first_eval_batch.reshape((1, n_input, n_features))

for i in range(len(test)):
    
    # get the prediction value for the first batch
    current_pred = model.predict(current_batch)[0]
    
    # append the prediction into the array
    test_predictions.append(current_pred) 
    
    # use the prediction to update the batch and remove the first value
    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
```

```python
true_predictions = scaler.inverse_transform(test_predictions)
```
